```python
import numpy as np
import pandas as pd


# converting csv file to dataframe
file = pd.read_csv('/Users/jeffreybiondi/Desktop/Jeffums Portfolio/moviesPythonData.csv', index_col = 0)

file
```

```python
# Exploring United States in the Data

titleCount = file['title'].str.contains('United States').sum()

print ('There are ' + str(titleCount) + ' movies that mention United States in the title.')

descCount = file['description'].str.contains('United States').sum()

print('There are ' + str(descCount) + ' movies that mention United States in the description.')

countryCount = file['country'].str.contains('United States').sum()

print('There are ' + str(countryCount) + ' movies that are from the United States.')
```

```python
# A Funtion That Will Search for a Director

def searchDirectors(directorName):
    myDirectors = file['director'].str.upper()
    # this makes all values in director key column UPPER
    # print(type(myDirectors))
    directorCount =  myDirectors.str.contains(directorName.upper()).sum()
    # counts the number of rows containing directorName
    if directorCount > 0:
      print('There are', directorCount, 'directors named', directorName, 'in this dataset.')
    else:
      print('No directors were found with the name:', directorName)

searchDirectors("pAuL")
# Using the fuction to search for input()
```

```python
# Finding the Most Common Words in the Longest Duration Movies

from collections import Counter

movieDurations = file[['title','duration_min','description']].dropna()
# print(movieDurations)
# movies' durations and description (no sort, remove NaN)

durationMax = movieDurations.max()
print('The longest movie in minutes is:',durationMax["duration_min"])

top10 = movieDurations.sort_values(by = 'duration_min', ascending = False).head(10)
print(top10)
# top 10 movies sorted by largest duration

print('\n')

print('The 20 most common words in the description for the top 10 longest movies:')
words = Counter(" ".join(top10['description']).lower().split()).most_common(20)
print(words)
# splits description and counts the number of most common words (lowercase)
```
